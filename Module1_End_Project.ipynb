Mini Project: Banking System Using Python
Project Description
Your task is to Banking System using Python that simulates basic banking operations.
Mandatory Requirements
Your program must implement the following features:

**Account Login**

● Allow users to log in using an account number/username and PIN/password
● Validate user credentials
● Display appropriate error messages for invalid login attempts

**Amount Deposit**

● Allow logged-in users to deposit money
● Deposit amount must be greater than zero
● Update and display the current balance after deposit

**Amount Withdrawal**

● Allow users to withdraw money
● Withdrawal amount must be greater than zero● Withdrawal amount should not exceed the available balance
● Display the remaining balance after withdrawal

```python
#Sample login credentials
sample_account_number = "11223344"
sample_pin ="11223344"
sample_account_balance = 10000.00

#Account Login
print("**********WELCOME TO ATHIRA'S BANK SYSTEM**********")
in_account_number = input("Please enter your account number : ")
in_pin = input("Please enter your pin : ")
#Validate user credentials
if in_account_number == sample_account_number and in_pin == sample_pin:
  while True:
    print("\nLogin Successful")
    print("\n Banking Menu")
    print("1. Amount Deposit")
    print("2. Amount Withdrawel")
    print("3. Check Balance")
    print("4. Exit")

    in_option = input("Please enter your option : ")
    # Amount Deposit
    if in_option == "1":
        deposit_amount = float(input("Please enter the amount to deposit : "))
        if deposit_amount > 0:
            sample_account_balance += deposit_amount
            print("\nAmount Deposited to your Account")
            print(f"Current Balance : {sample_account_balance}")
        else:
            print("\nInvalid Amount. Amount must be greater than zero")
    # Amount Withdrawel
    elif in_option == "2":
        withdraw_amount = float(input("Please enter the amount to withdraw : "))
        if withdraw_amount <= 0:
            print("\nInvalid Amount. Amount must be greater than zero")
        elif withdraw_amount > sample_account_balance:
            print("\nInsufficient Balance")
        else:
            sample_account_balance -= withdraw_amount
            print("\nAmount Withdrawn from your Account")
            print(f"Current Balance : {sample_account_balance}")
    # Check Balance
    elif in_option == "3":
        print(f"\nCurrent Balance : {sample_account_balance}")
    # Exit
    elif in_option == "4":
        print("\nThank you for using our services. Have a nice day")
        break
    else:
        print("\nInvalid Option. Please select a valid option")
else:
    print("\nInvalid Account Number or Pin. \nAccess Denied")
```
