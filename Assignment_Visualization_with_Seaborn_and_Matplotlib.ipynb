**1. Countplot: Passenger Count by Embarked Port**

Create a countplot to display the number of passengers who boarded from each
port (embarked), separated by sex

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset from Seaborn
titanic = sns.load_dataset("titanic")
#titanic

# Create countplot
sns.countplot(
    data=titanic,
    x="embarked",
    hue="sex",
    palette={"male": "blue", "female": "pink"}
)

# Labels and title
plt.xlabel("Embarked Port")
plt.ylabel("Passenger Count")
plt.title("Countplot:Passenger Count by Embarked Port and Sex")

# Show plot
plt.show()
```

**Insights**

1.Southampton (S) had the highest number of passengers, with males significantly outnumbering females.

2.Male passengers dominate across all embarkation ports, though the gap is smaller at Cherbourg (C) and Queenstown (Q).

3.Queenstown (Q) had the lowest passenger count and the most balanced gender
distribution.


**2. Swarmplot: Age Distribution across Classes**

 Create a swarmplot to visualize the distribution of age across passenger classes (pclass), categorized by sex



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset from Seaborn
titanic = sns.load_dataset("titanic")
#titanic

# Create swarmplot
sns.swarmplot(
    data=titanic,
    x="pclass",
    y="age",
    hue="sex",
    palette={"male": "black", "female": "red"},
    size=4
)

# Labels and title
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.title("Swarmplot:Age Distribution across Passenger Classes by Sex")

# Show plot
plt.show()
```

**Insights**

1.First-class passengers tend to be older, with a wider age range compared to other classes.

2.Third-class passengers are generally younger, showing a higher concentration of children and young adults.

3.Male passengers show greater age spread across all classes, while females are more tightly clustered in each class.







**3. Pie Chart: Gender Distribution**

Create a pie chart using Matplotlib to show the percentage of male vs. female
passengers. - Add percentage labels using autopct - Use explode for one slice to
highlight it.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset from Seaborn
titanic = sns.load_dataset("titanic")
#titanic

gender_counts = titanic['sex'].value_counts()

plt.figure()
plt.pie(
    gender_counts.values,
    labels=gender_counts.index,
    colors= ["red","green"],
    explode=(0.2, 0),
    autopct='%1.1f%%',
    shadow=True,
    startangle=90
)

plt.title("Pie Chart:Gender Distribution")
plt.axis('equal')
plt.show()
```

**Insights**

1.Male passengers constitute the majority of the onboard population (approximately 64.8%), indicating that nearly two-thirds of the passengers were men.

2.Female passengers account for about 35.2% of the total, which shows a substantial but significantly smaller representation compared to males.



**4. Heatmap: Correlation Matrix**

Create a heatmap to show correlations between age, fare, pclass, sibsp, parch,
and survived. (Use seaborn.heatmap with annotations)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset from Seaborn
titanic = sns.load_dataset("titanic")
#titanic

# Select relevant numeric columns
data = titanic.select_dtypes(include='number')
# Compute correlation matrix
corr_matrix = data.corr()

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(
    corr_matrix,
    annot=True,       # show correlation values
    cmap='coolwarm',  # color map
    fmt=".2f",        # 2 decimal places
    linewidths=0.5
)

plt.title("Heatmap:Correlation Matrix")
plt.show()
```

**Insights**


1.Passengers in higher classes paid more and survived more often.

2.People traveling with family had a slightly better chance of survival, but the effect was small.

3.Age did not strongly affect survival, even though older passengers were usually in higher classes.










**5. Violin Plot: Fare by Class and Gender**

Create a violin plot to visualize fare distribution across classes, separated by
gender

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset from Seaborn
titanic = sns.load_dataset("titanic")

# Create violin plot
sns.violinplot(
    data=titanic,
    x="pclass",
    y="fare",
    hue="sex",
    split=True,
    palette="Set3"
)

# Labels and title
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.title("Violin Plot: Fare by Class and Gender")

plt.show()
```

**Insights**

1.First-class passengers paid much higher fares, with a wider spread.

2.Third-class fares are low and tightly clustered.

3.Female passengers generally paid slightly higher fares, especially in first class.









**6. Scatter Plot: Age vs Fare**

Create a scatter plot to explore the relationship between age and fare, colored by survival status.

Use a seaborn scatterplot with hue='survived'. Each plot must have titles, labels,and legends where needed.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
titanic = sns.load_dataset("titanic")

# Create scatter plot
sns.scatterplot(
    data=titanic,
    x="age",
    y="fare",
    hue="survived"
)

# Labels and title
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Scatter Plot: Age vs Fare")

plt.show()
```

**Insights**

1.Passengers who paid higher fares were more likely to survive.

2.Survivors are spread across many age groups, showing age alone did not decide survival.

3.Lower-fare passengers show a higher concentration of non-survivors.
