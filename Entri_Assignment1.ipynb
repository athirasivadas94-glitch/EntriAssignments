**1)Topic :List**
1. Create a list of 5 random numbers and print the list.
2. Insert 3 new values to the list and print the updated list.
3. Try to use a for loop to print each element in the list.

```python
#Using random
#Create a list of 5 random numbers and print the list.
import random
randomlist = []
randomlist = [random.randint(1, 100) for _ in range(5)]
print(randomlist)

#Insert 3 new values to the list and print the updated list.
for i in range(0,3):
  n = random.randint(1,50)
  randomlist.append(n)
print(randomlist)

#Try to use a for loop to print each element in the list.

for i in randomlist:
  print(i)
```

```python
#Create a list of 5 random numbers and print the list.
mylist = [2,4,6,8,10]
print(mylist)
```

```python
#Insert 3 new values to the list and print the updated list.
mylist.append(12)#insert 12 using append
mylist.insert(6,14)#insert 14 at index 6
newlist=[16]#create a new list with one element
mylist.extend(newlist)#extending mylist with newlist
print(mylist)
```

```python
#Try to use a for loop to print each element in the list.
for i in mylist:
  print(i)
```
